```python
"""
NMR Predictor Pipeline
======================
A computational workflow for predicting NMR chemical shifts from molecular structures.
Combines RDKit for conformer generation and PySCF for quantum chemical NMR calculations.
"""

from rdkit import Chem
from rdkit.Chem import Draw, rdDistGeom, rdForceFieldHelpers, AllChem
import py3Dmol
import numpy as np
from pathlib import Path
from typing import List, Dict, Tuple, Optional, Union

# Configuration constants
PDBFILE: str = 'cyclohexane.pdb'
"""str: Path to input PDB file containing the molecular structure."""

RANDOM_SEED: int = 45
"""int: Random seed for reproducible conformer generation."""
```

```python
## Phase 1
```

```python
def get_geom(pdbfile: str) -> Tuple[Chem.Mol, int, object]:
    """
    Generate a 3D structure from a PDB file with explicit hydrogens and force field optimization.
    
    Parameters
    ----------
    pdbfile : str
        Path to the input PDB file
    
    Returns
    -------
    Tuple[Chem.Mol, int, object]
        - RDKit molecule object with explicit hydrogens and optimized geometry
        - Conformer ID of the embedded structure
        - PIL Image of the molecule for visualization
    
    Notes
    -----
    Uses MMFF force field for geometry optimization after initial embedding.
    """
    implicit_hs = Chem.MolFromPDBFile(pdbfile)
    explicit_hs = Chem.AddHs(implicit_hs)
    conf_id = rdDistGeom.EmbedMolecule(explicit_hs, maxAttempts=2000, randomSeed=RANDOM_SEED)
    rdForceFieldHelpers.MMFFOptimizeMolecule(explicit_hs)
    img = Draw.MolToImage(explicit_hs)
    return explicit_hs, conf_id, img

def export_molecule(structure: Chem.Mol, file_type: str = 'pdb') -> None:
    """
    Export molecular structure to various file formats.
    
    Parameters
    ----------
    structure : Chem.Mol
        RDKit molecule object to export
    file_type : str, optional
        Export format: 'pdb', 'mol', or 'xyz' (default: 'pdb')
    
    Raises
    ------
    ValueError
        If an unsupported file type is specified
    """
    filename = f'{Path(PDBFILE).stem}.{file_type}'
    with open(filename, 'w') as f:
        if file_type == 'pdb':
            f.write(Chem.MolToPDBBlock(structure))
        elif file_type == 'mol':
            f.write(Chem.MolToMolBlock(structure))
        elif file_type == 'xyz':
            f.write(Chem.MolToXYZBlock(structure))
        else:
            raise ValueError(f"Unsupported file type: {file_type}. Use 'pdb', 'mol', or 'xyz'.")
            
def view_structure(structure: Chem.Mol) -> None:
    """
    Display 3D molecular structure using py3Dmol in Jupyter notebook.
    
    Parameters
    ----------
    structure : Chem.Mol
        RDKit molecule object to visualize
    
    Notes
    -----
    Renders an interactive 3D viewer with stick representation.
    """
    mb = Chem.MolToMolBlock(structure)
    viewer = py3Dmol.view(width=500, height=400)
    viewer.addModel(mb, 'mol')
    viewer.setStyle({'stick': {}})
    viewer.show()
        
structure, conf_id, image = get_geom(PDBFILE)
```

```python
def get_atom_table(structure: Chem.Mol) -> List[Dict]:
    """
    Extract atomic properties into a structured table.
    
    Parameters
    ----------
    structure : Chem.Mol
        RDKit molecule object
    
    Returns
    -------
    List[Dict]
        List of dictionaries containing atom properties:
        - idx: atom index
        - element: atomic symbol
        - is_aromatic: whether atom is in aromatic ring
        - in_ring: whether atom is in any ring
    """
    atoms = []
    for atom in structure.GetAtoms():
        atoms.append({
            'idx': atom.GetIdx(),
            'element': atom.GetSymbol(),
            'is_aromatic': atom.GetIsAromatic(),
            'in_ring': atom.IsInRing()        
        })
    return atoms

def get_bond_table(structure: Chem.Mol) -> List[Dict]:
    """
    Extract bond properties into a structured table.
    
    Parameters
    ----------
    structure : Chem.Mol
        RDKit molecule object
    
    Returns
    -------
    List[Dict]
        List of dictionaries containing bond properties:
        - atom_i: index of first atom
        - atom_j: index of second atom
        - bond_order: bond order as float
        - in_ring: whether bond is in a ring
    """
    bonds = []
    for bond in structure.GetBonds():
        bonds.append({
            'atom_i': bond.GetBeginAtomIdx(),
            'atom_j': bond.GetEndAtomIdx(),
            'bond_order': bond.GetBondTypeAsDouble(),
            'in_ring': bond.IsInRing()
        })
    return bonds

def get_equiv_groups(structure: Chem.Mol) -> Dict[int, List[int]]:
    """
    Identify symmetry-equivalent atoms using canonical ranking.
    
    Parameters
    ----------
    structure : Chem.Mol
        RDKit molecule object
    
    Returns
    -------
    Dict[int, List[int]]
        Dictionary mapping rank to list of atom indices with that rank.
        Atoms with the same rank are symmetry-equivalent.
    """
    ranks = list(Chem.CanonicalRankAtoms(structure, breakTies=False))
    groups = {}
    for idx, rank in enumerate(ranks):
        if rank not in groups:
            groups[rank] = []
        groups[rank].append(idx)
    return groups

atom_table = get_atom_table(structure)
bond_table = get_bond_table(structure)
equiv_groups = get_equiv_groups(structure)
```

```python
## Phase 2
```

```python
def prune_conformers(structure: Chem.Mol, conf_ids: List[int], rmsd_threshold: float = 0.3) -> List[int]:
    """
    Remove duplicate conformers based on RMSD similarity.
    
    Parameters
    ----------
    structure : Chem.Mol
        RDKit molecule with multiple conformers
    conf_ids : List[int]
        List of conformer IDs to prune
    rmsd_threshold : float, optional
        RMSD threshold in Angstroms for considering conformers as duplicates (default: 0.3)
    
    Returns
    -------
    List[int]
        Filtered list of unique conformer IDs
    """
    keep = []
    for i in conf_ids:
        duplicate = False
        for j in keep:
            rmsd = AllChem.GetBestRMS(structure, structure, i, j)
            if rmsd < rmsd_threshold:
                duplicate = True
                break
        if not duplicate:
            keep.append(i)
    return keep

def boltzmann_weights(energies_kcal: List[float], temperature: float = 298.15) -> np.ndarray:
    """
    Calculate Boltzmann weights from relative energies.
    
    Parameters
    ----------
    energies_kcal : List[float]
        List of energies in kcal/mol
    temperature : float, optional
        Temperature in Kelvin (default: 298.15)
    
    Returns
    -------
    np.ndarray
        Normalized Boltzmann weights (sum to 1)
    
    Notes
    -----
    Uses R = 0.001987 kcal/(mol·K) as the gas constant.
    """
    R = 0.001987  # kcal/mol/K
    energies = np.array(energies_kcal)
    energies -= energies.min()  # Shift to zero minimum
    weights = np.exp(-energies / (R * temperature))
    weights /= weights.sum()
    return weights

# Generate and optimize conformers
conf_ids = rdDistGeom.EmbedMultipleConfs(structure, numConfs=100, randomSeed=RANDOM_SEED)
energies = []
for conf_id in conf_ids:
    rdForceFieldHelpers.MMFFOptimizeMolecule(structure, confId=conf_id)
    ff = rdForceFieldHelpers.MMFFGetMoleculeForceField(
            structure,
            rdForceFieldHelpers.MMFFGetMoleculeProperties(structure),
            confId=conf_id)
    energies.append((conf_id, ff.CalcEnergy()))

# Sort and prune conformers
energies_sorted = sorted(energies, key=lambda x: x[1])
sorted_conf_ids = [conf_id for conf_id, _ in energies_sorted]

keep = prune_conformers(structure, sorted_conf_ids)
kept_energies = [e for conf_id, e in energies_sorted if conf_id in keep]
weights = boltzmann_weights(kept_energies)
weighted_conformers = list(zip(keep, weights))

# Export unique conformers
for conf_id, weight in weighted_conformers:
    with open(f'conformer_{conf_id}.xyz', 'w') as f:
        f.write(Chem.MolToXYZBlock(structure, confId=conf_id))
```

```python
# import warnings
# warnings.filterwarnings('ignore')
from pyscf import gto, dft, scf
from pyscf.prop.nmr import rhf as rhf_nmr
import numpy as np

# ── Monkey-patch for PySCF bug: get_vxc_giao passes blksize=ngrids to
# block_loop, but ngrids is not guaranteed to be a multiple of BLKSIZE=256.
# This rounds any externally-supplied blksize down to the nearest valid
# multiple before block_loop sees it, preventing the AssertionError.
from pyscf.dft import numint as _numint
_orig_block_loop = _numint.NumInt.block_loop

def _patched_block_loop(self, mol, grids, nao, deriv=0, max_memory=2000,
                         non0tab=None, blksize=None, buf=None):
    _BLKSIZE = _numint.BLKSIZE
    if blksize is not None and blksize % _BLKSIZE != 0:
        blksize = max(_BLKSIZE, (blksize // _BLKSIZE) * _BLKSIZE)
        buf = None  # let block_loop re-allocate to match corrected blksize
    return _orig_block_loop(self, mol, grids, nao, deriv, max_memory,
                             non0tab=non0tab, blksize=blksize, buf=buf)

_numint.NumInt.block_loop = _patched_block_loop
# ── end monkey-patch ──

def calculate_nmr(xyz_file, basis='6-31G*', functional='b3lyp'):
    """
    Calculate NMR shielding tensors from an XYZ file using DFT (RKS).
    """
    with open(xyz_file) as f:
        lines = f.readlines()
        xyz_block = ''.join(lines[2:])

    mol = gto.Mole(atom=xyz_block, basis=basis, charge=0, spin=0, verbose=0)
    mol.build()

    print(f"Computing NMR for {xyz_file}")
    print(f"Functional: {functional}, Basis: {basis}")
    print(f"  {mol.natm} atoms, {mol.nao} AOs, {mol.nelectron} electrons")

    def _build_scf(level_shift=0.0, max_cycle=200, conv_tol=1e-9):
        mf = dft.RKS(mol)
        mf.xc = functional
        mf.grids.level = 4
        mf.level_shift = level_shift
        mf.max_cycle = max_cycle
        mf.conv_tol = conv_tol
        mf.kernel()
        return mf

    mf = _build_scf()
    if not mf.converged:
        print("    SCF not converged, retrying with level shift...")
        mf = _build_scf(level_shift=0.2)

    from pyscf.prop.nmr import rks as rks_nmr
    nmr_calc = rks_nmr.NMR(mf)
    return nmr_calc.kernel()

def isotropic_shielding(shielding_tensors):
    return {i: np.trace(shielding_tensors[i]) / 3 for i in range(len(shielding_tensors))}

def average_shieldings(all_shieldings, weighted_conformers, n_atoms):
    """Return dict {atom_idx: Boltzmann‑averaged isotropic shielding}."""
    averaged = {}
    for atom_idx in range(n_atoms):
        sigma = 0.0
        for conf_id, weight in weighted_conformers:
            sigma += weight * all_shieldings[conf_id][atom_idx]
        averaged[atom_idx] = sigma
    return averaged

all_shieldings = {}
for conf_id, weight in weighted_conformers:
    xyz_file = f'conformer_{conf_id}.xyz'
    tensors = calculate_nmr(xyz_file)
    all_shieldings[conf_id] = isotropic_shielding(tensors)
    
n_atoms = structure.GetNumAtoms()
avg_iso = average_shieldings(all_shieldings, weighted_conformers, n_atoms)
print("Averaged isotropic shieldings (ppm):")
for idx, val in avg_iso.items():
    print(f"{idx:2d} {structure.GetAtomWithIdx(idx).GetSymbol()} : {val:.3f}")

# try:
#     shielding = calculate_nmr(xyz_file, method='dft')
#     print("DFT calculation successful!")
# except Exception as e:
#     print(f"DFT failed ({e}), using RHF...")
#     shielding = calculate_nmr(xyz_file, method='rhf')
```

```python
from rdkit.Chem import AllChem

# Build TMS from SMILES
tms_smiles = "C[Si](C)(C)C"
tms_mol = Chem.MolFromSmiles(tms_smiles)
tms_mol = Chem.AddHs(tms_mol)
AllChem.EmbedMolecule(tms_mol, randomSeed=RANDOM_SEED)
AllChem.MMFFOptimizeMolecule(tms_mol)

# Write temporary XYZ file
tms_xyz = "tms.xyz"
with open(tms_xyz, 'w') as f:
    f.write(Chem.MolToXYZBlock(tms_mol))
```

```python
tms_xyz = "tms.xyz"
with open(tms_xyz, 'w') as f:
    f.write(Chem.MolToXYZBlock(tms_mol))

# Compute TMS shieldings (use same method as for conformers, here RHF)
tms_tensors = calculate_nmr(tms_xyz, basis='6-31G*')
tms_iso = isotropic_shielding(tms_tensors)   # dict atom_idx -> shielding

# Average reference values for carbon and hydrogen (all equivalent in TMS)
ref_C = np.mean([tms_iso[i] for i, atom in enumerate(tms_mol.GetAtoms()) if atom.GetSymbol() == 'C'])
ref_H = np.mean([tms_iso[i] for i, atom in enumerate(tms_mol.GetAtoms()) if atom.GetSymbol() == 'H'])

print(f"\nTMS reference:  C = {ref_C:.3f} ppm,  H = {ref_H:.3f} ppm")
```

```python
chemical_shifts = {}
for idx, atom in enumerate(structure.GetAtoms()):
    if atom.GetSymbol() == 'C':
        chemical_shifts[idx] = ref_C - avg_iso[idx]
    elif atom.GetSymbol() == 'H':
        chemical_shifts[idx] = ref_H - avg_iso[idx]
    else:
        chemical_shifts[idx] = np.nan   # ignore other nuclei for now

print("\nPredicted chemical shifts (ppm):")
for idx in sorted(chemical_shifts):
    if not np.isnan(chemical_shifts[idx]):
        print(f"{idx:2d} {structure.GetAtomWithIdx(idx).GetSymbol()} : {chemical_shifts[idx]:.3f}")
```
